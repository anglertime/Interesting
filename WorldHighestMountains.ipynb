```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
mount = pd.read_csv('C:\Users\Ming\My Documents\Notebooks\MountainsMing.csv')
```

```python
mount['first'] = mount.Coordinates.apply(lambda x: x.split()[0][0:2] + '.' + x.split()[0][3:5]).astype('float')
```

```python
mount['second'] = mount.Coordinates.apply(lambda x: x.split()[1][0:2] + '.' + x.split()[1][3:5]).astype('float')
```

```python
mount.head()
```

```python
%matplotlib inline
plt.figure(1, figsize=(20,10))
m = Basemap(projection='merc',
             llcrnrlat=-60,
             urcrnrlat=70,
             llcrnrlon=-180,
             urcrnrlon=180,
             lat_ts=0,
             resolution='h')

m.drawcoastlines()

x, y = m(mount['second'].tolist(), mount['first'].tolist())

m.plot(x, y, 
            'o', c='r', lw=0, alpha=1, zorder=5 , ms = 50        # marker
            )
plt.title("World's_Highest_Mountains")
plt.show()

```
